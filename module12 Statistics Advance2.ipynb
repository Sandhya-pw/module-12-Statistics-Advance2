Theoretical Questions

```python
# 1. What is hypothesis testing in statistics?
'''Hypothesis testing in statistics is a method used to make decisions or inferences about population parameters based on sample data.'''
```

```python
# 2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
''' The null hypothesis (H₀) and the alternative hypothesis (Ha) are two opposing statements used in hypothesis testing.
Null Hypothesis (H₀): Assumes no effect or difference; it's the default claim tested.
Example:
H₀: The average blood pressure of a population = 120 mmHg
(There is no change or effect.)

Alternative Hypothesis (Ha):Claims there is an effect or difference; it's what you want to prove.
Example:
Ha: The average blood pressure ≠ 120 mmHg
(There is a change or effect.)'''
```

```python
# 3. What is the significance level in hypothesis testing, and why is it important?
'''Significance level (α), also called the Type I error rate, is the probability of rejecting the null hypothesis when it is actually true.
Importance:
It sets the threshold for decision-making in hypothesis testing.
Common values: 0.05 (5%), 0.01 (1%)
It helps control the risk of making a false positive conclusion.'''
```

```python
# 4. What does a P-value represent in hypothesis testing?
'''P-value is the probability value calculated from a statistical test (based on the test statistic like a z-score).
It is used in hypothesis testing to decide whether to reject the null hypothesis.'''
```

```python
# 5. How do you interpret the P-value in hypothesis testing?
'''Interpretion:
Low P-value (≤ α): Strong evidence against H₀ → Reject H₀
High P-value (> α): Weak evidence against H₀ → Do not reject H₀'''
```

```python
# 6. What are Type 1 and Type 2 errors in hypothesis testing?
'''Type 1 error: is the rejection of the null hypothesis when it is actully true
   Type 2 error: is the failure to reject the null hypothesis that is actully false.'''
```

```python
# 7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
 '''one-tail test:
   in this type of test there is rejection region at one side of distribution ( where Ha: μ > μ₀ or H₁: μ < μ₀).
   i.e Tests for effect in one direction only.

   two-tail test: in this type of test there is rejection region at both side of distribution( Ha: μ ≠ μ₀)
   i.e Tests for effect in both directions'''
```

```python
# 8. What is the Z-test, and when is it used in hypothesis testing?
'''A Z-test is used in hypothesis testing to compare sample and population means
when the population standard deviation is known and the sample size is large.(n >= 30)
It helps determine if a sample result is significantly different from a population value.'''
```

```python
# 9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
'''Calculation of z score = (mean of sample - population mean)/(sd of population/square root of sample size).

It represents,how many standard errors the sample mean is from the population mean under the null hypothesis.
A high absolute Z-score indicates the sample mean is far from the population mean.
It helps determine the P-value, which is used to decide whether to reject H₀. '''
```

```python
# 10. What is the T-distribution, and when should it be used instead of the normal distribution?
'''The T-distribution is a bell-shaped, symmetric probability distribution used in statistics/hypothesis testing when estimating population parameters.
 It is used when the sample size is small (n < 30) and the population standard deviation is unknown.''
```

```python
# 11. What is the difference between a Z-test and a T-test?
'''A Z-test is typically used when the population standard deviation is known and the sample size is large (usually over 30).
It assumes that the data follows a normal distribution and relies on the standard normal distribution (Z-distribution) to calculate probabilities.
Shape of Curve is narrower (less variability)

A T-test, on the other hand, is used when the population standard deviation is unknown, and the sample size is small (under 30).
It uses the t-distribution.
Shape of Curve is wider tails (more variability)'''
```

```python
# 12. What is the T-test, and how is it used in hypothesis testing?
'''A T-test is a statistical test used to compare means to determine if they are significantly different.
It checks whether the difference between sample mean and population mean (or between two sample means) is statistically significant.

Used when the sample size is small (n < 30) and the population standard deviation is unknown.'''
```

```python
# 13. What is the relationship between Z-test and T-test in hypothesis testing?
''' The Z-test and T-test are both used in hypothesis testing to determine if there is a significant difference between sample data and population data.

Difference:
Z-test is used when the sample size is large (typically n > 30) or when the population standard deviation is known.
It assumes the data follows a normal distribution.

T-test is used when the sample size is small (n < 30) or when the population standard deviation is unknown.
It uses the t-distribution, which adjusts for the smaller sample size.

As the sample size increases, the T-distribution approaches the normal distribution (Z-test).
Therefore, for large samples, the T-test can be approximated by a Z-test.'''
```

```python
# 14. What is a confidence interval, and how is it used to interpret statistical results?
'''A confidence interval (CI) is a range of values that is used to estimate a population parameter (such as a population mean or proportion).
It provides an interval within which the true value of the parameter is likely to lie, with a certain level of confidence.'''
```

```python
# 15. What is the margin of error, and how does it affect the confidence interval?
'''The margin of error shows how much the survey result could be off from the true value.
It affects the confidence interval by making it wider or narrower.

A bigger margin means a wider range, showing less certainty,
while a smaller margin means a narrower range, showing more confidence in the result.'''
```

```python
# 16. How is Bayes' Theorem used in statistics, and what is its significance?
'''Bayes' Theorem helps us update the probability of an event based on new information.
It combines what we already know (prior probability) and new evidence (likelihood) to give a more accurate prediction (posterior probability).

Significance:
1.Updating Beliefs: It helps in revising the probability of an event as new information becomes available.
2.Decision Making: It's used in fields like medical diagnosis, spam filtering, and machine learning to make better predictions based on prior knowledge.
3.Incorporating Uncertainty: Bayes' Theorem incorporates both existing knowledge (prior probability) and new evidence (likelihood),
making it powerful for uncertain situations.'''
```

```python
# 17. What is the Chi-square distribution, and when is it used?
'''The Chi-square distribution is a statistical distribution that is used to test hypotheses about categorical data.
USE:
Goodness-of-Fit Test: To see if a sample data matches a population with a specific distribution.
Test of Independence: To test if two categorical variables are independent of each other (e.g., in a contingency table).
Test for Homogeneity: To test if different groups share the same distribution.'''
```

```python
# 18. What is the Chi-square goodness of fit test, and how is it applied?
'''The Chi-square goodness of fit test checks if observed data matches expected frequencies.
It compares the actual data to the expected data to see if there’s a significant difference.

If the Chi-square value is higher than the critical value, reject the null hypothesis, indicating a significant difference.'''
```

```python
# 19. What is the F-distribution, and when is it used in hypothesis testing?
'''The F-distribution is used in statistics to compare variances between two or more groups. It helps in hypothesis testing, especially in:

ANOVA: To check if there are significant differences in means across multiple groups.
Variance comparison: To test if two groups have the same variance.
Regression analysis: To test the overall significance of a model.

A large F value suggests significant differences between groups.'''
```

```python
# 20. What is an ANOVA test, and what are its assumptions?
'''ANOVA (Analysis of Variance) is a statistical test used to compare the means of three or more groups
 to determine if at least one group mean is significantly different from the others.

Assumptions of ANOVA:
Independence: The samples from each group should be independent of each other.
Normality: The data in each group should be approximately normally distributed.
absense of outliers.
Homogeneity of variances (Homoscedasticity): The variances within each group should be roughly equal.

If these assumptions are met, ANOVA helps to determine whether there is a statistically significant difference in group means.'''
```

```python
# 21. What are the different types of ANOVA tests?
'''Types of ANOVA:

1.One-Way ANOVA – One factor, different groups, 1 independent variable, 1 dependent variable.
eg. Compare average marks of students in 3 schools.

2.Repeated Measures ANOVA – Same subjects tested multiple times.
eg. Measure blood pressure of patients before, during, and after treatment.

3.Factorial ANOVA – Two factors, tests interaction, 2 or more independent variables, tests interaction effects.
eg. Check how teaching method and gender affect scores.

Dependent variable = What we measure
Independent variable = What we change or compare
levels are always part of the independent variable.'''
```

```python
# 22. What is the F-test, and how does it relate to hypothesis testing?
'''The F-test is a statistical test used to compare variances across multiple groups to determine
 if they are significantly different from each other.

The F-statistic is calculated as the ratio of variance between groups to variance within groups.
If the F-statistic is large, we reject the null hypothesis, indicating a significant difference between group means.'''
```

```python
# 1 Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.
def one_sample_z_test(sample, pop_mean, pop_std):
    n = len(sample)
    sample_mean = np.mean(sample)
    z = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
    p_value = 2 * (1 - norm.cdf(abs(z)))
    print(f"Z-score: {z:.2f}, P-value: {p_value:.4f}")
    return z, p_value

sample = np.random.normal(100, 15, 50)
one_sample_z_test(sample, pop_mean=100, pop_std=15)
```

```python
# 3.Implement a one-sample Z-test using Python to compare the sample mean with the population mean.
import numpy as np
import scipy.stats as stats

# Sample data and population parameters
sample_data = [45, 50, 52, 48, 49]
population_mean = 50
population_std = 5

# Z-test calculation
sample_mean = np.mean(sample_data)
z = (sample_mean - population_mean) / (population_std / np.sqrt(len(sample_data)))

# p-value for two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

print(f"Z-statistic: {z}")
print(f"p-value: {p_value}")

# Interpretation
print("Reject H0" if p_value < 0.05 else "Fail to reject H0")
```

```python
# 4.Perform a two-tailed Z-test using Python and visualize the decision region on a plot.
def visualize_z_test(z, alpha=0.05):
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    z_critical = norm.ppf(1 - alpha / 2)

    plt.plot(x, y, label='Standard Normal Distribution')
    plt.fill_between(x, y, where=(x <= -z_critical) | (x >= z_critical), color='red', alpha=0.3, label='Rejection Region')
    plt.axvline(z, color='blue', linestyle='--', label=f'Z-score = {z:.2f}')
    plt.title("Two-Tailed Z-test Decision Region")
    plt.legend()
    plt.grid(True)
    plt.show()

visualize_z_test(z)
```

```python
# 5.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
def visualize_type_errors(mu0=0, mu1=1, sigma=1, n=30, alpha=0.05):
    se = sigma / np.sqrt(n)
    x = np.linspace(-4, 4, 1000)

    z_crit = norm.ppf(1 - alpha / 2)
    x_cutoff = mu0 + z_crit * se

    plt.plot(x, norm.pdf(x, mu0, se), label='H0', color='blue')
    plt.plot(x, norm.pdf(x, mu1, se), label='H1', color='green')

    plt.fill_between(x, norm.pdf(x, mu0, se), where=(x > x_cutoff), color='red', alpha=0.3, label='Type I Error (α)')
    plt.fill_between(x, norm.pdf(x, mu1, se), where=(x < x_cutoff), color='orange', alpha=0.3, label='Type II Error (β)')

    plt.legend()
    plt.grid(True)
    plt.title("Type I and Type II Errors")
    plt.show()

visualize_type_errors()
```

```python
# 9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import numpy as np
from scipy import stats
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    sem = stats.sem(data)
    n = len(data)
    margin = t.ppf((1 + confidence) / 2., n-1) * sem
    ci = (mean - margin, mean + margin)
    print(f"Mean = {mean:.2f}, 95% CI = {ci}")
    return ci

data = np.random.normal(70, 10, 50)
confidence_interval(data)
```

```python
# 10.Write a Python program to calculate the margin of error for a given confidence level using sample data.
import numpy as np
from scipy.stats import t, sem

data = np.random.normal(100, 15, 50)
confidence = 0.95

mean = np.mean(data)
standard_error = sem(data)
t_value = t.ppf((1 + confidence) / 2, df=len(data)-1)

margin_error = t_value * standard_error

# Print both margin of error and confidence interval
print("Margin of Error:", margin_error)
print(f"95% Confidence Interval: ({mean - margin_error:.2f}, {mean + margin_error:.2f})")
```

```python
# 11.Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
P_disease = 0.01
P_no_disease = 1 - P_disease

P_positive_given_disease = 0.99
P_positive_given_no_disease = 0.05

P_positive = (P_positive_given_disease * P_disease) + (P_positive_given_no_disease * P_no_disease)# Total probability of testing positive

P_disease_given_positive = (P_positive_given_disease * P_disease) / P_positive # Bayes' Theorem

print("Probability of having the disease given a positive test:", round(P_disease_given_positive, 4))
```

```python
# 12.Perform a Chi-square test for independence between two categorical variables in Python.
import pandas as pd
from scipy.stats import chi2_contingency

data = [[20, 15],
        [30, 35]]

table = pd.DataFrame(data, columns=['Group A', 'Group B'], index=['Yes', 'No'])
chi2, p, dof, expected = chi2_contingency(table) # Perform Chi-square test for independence

print("Chi-square value:", chi2)
print("p-value:", p)
print("Degrees of freedom:", dof)
print("Expected frequencies:\n", expected)


if p < 0.05:  # Interpretation
    print("Variables are dependent (reject H0)")
else:
    print("Variables are independent (fail to reject H0)")
```

```python
# 14.Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.
from scipy.stats import chisquare

observed = [18, 22, 20, 25, 15] # Observed frequencies
expected = [20, 20, 20, 20, 20] # Expected frequencies (e.g., equal distribution)
chi2_stat, p_val = chisquare(f_obs=observed, f_exp=expected) # Perform Chi-square goodness-of-fit test

print("Chi-square statistic:", chi2_stat)
print("p-value:", p_val)

# Interpretation
if p_val < 0.05:
    print("Observed data does NOT fit the expected distribution (reject H0)")
else:
    print("Observed data fits the expected distribution (fail to reject H0)")
```
